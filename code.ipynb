```python
#importing needed libraries 
import matplotlib.pyplot as plt 
import pandas as pd 
from pathlib import Path
from matplotlib.gridspec import GridSpec
%matplotlib inline
```

```python
#Taking in the path to the files
file_accra_vec_tm = Path('csv_files/Accra_vectorcapacity_Tm.csv')
file_accra_vec_tn = Path('csv_files/Accra_vectorcapacity_Tn.csv')
file_accra_vec_tx = Path('csv_files/Accra_vectorcapacity_Tx.csv')
file_kum_vec_tm   = Path('csv_files/Kumasi_vectorcapacity_Tm.csv')
file_kum_vec_tn   = Path('csv_files/Kumasi_vectorcapacity_Tn.csv')
file_kum_vec_tx   = Path('csv_files/Kumasi_vectorcapacity_Tx.csv')
file_nav_vec_tm   = Path('csv_files/Navrongo_vectorcapacity_Tm.csv')
file_nav_vec_tn   = Path('csv_files/Navrongo_vectorcapacity_Tn.csv')
file_nav_vec_tx   = Path('csv_files/Navrongo_vectorcapacity_Tx.csv')
file_tem_vec_tm   = Path('csv_files/Tamale_vectorcapacity_Tm.csv')
file_tem_vec_tn   = Path('csv_files/Tamale_vectorcapacity_Tn.csv')
file_tem_vec_tx   = Path('csv_files/Tamale_vectorcapacity_Tx.csv')
```

```python
#Reading the datasets 

#Dataset for Accra
df_accra_vec_tm = pd.read_csv(file_accra_vec_tm, sep=',')
df_accra_vec_tn = pd.read_csv(file_accra_vec_tn, sep=',')
df_accra_vec_tx = pd.read_csv(file_accra_vec_tx, sep=',')

#Dataset for Kumasi
df_kum_vec_tm   = pd.read_csv(file_kum_vec_tm, sep=',')
df_kum_vec_tn   = pd.read_csv(file_kum_vec_tn, sep=',')
df_kum_vec_tx   = pd.read_csv(file_kum_vec_tx, sep=',')

#Dataset for Navorongo
df_nav_vec_tm   = pd.read_csv(file_nav_vec_tm, sep=',')
df_nav_vec_tn   = pd.read_csv(file_nav_vec_tn, sep=',')
df_nav_vec_tx   = pd.read_csv(file_nav_vec_tx, sep=',')



#Dataset for Tamale
df_tem_vec_tm   = pd.read_csv(file_tem_vec_tm, sep=',')
df_tem_vec_tn   = pd.read_csv(file_tem_vec_tn, sep=',')
df_tem_vec_tx   = pd.read_csv(file_tem_vec_tx, sep=',')

#month = df_tem_vec_tx.month
df_kum_vec_tx.head(4)
month = ['01','02','03','04','05','06','07','08','09','10','11','12']
```

```python
#Ploting graphs for Tn,Tm,Tx against P
fig, (ax1, ax2,ax3) = plt.subplots(3,figsize=(11, 16))
gs = GridSpec(nrows=2, ncols=2)

labels=['Accra','Navorongo','Tamale','Kumasi']

#ax1 = fig.add_subplot(gs[0, 0])

#ax2 = fig.add_subplot(gs[0, 1:])

#ax3 = fig.add_axes([1,0])
plt.suptitle('Fig 1. Monthly Survival Rate of Mosquito',fontsize=20)

# A plot of  Tn for each region 
ax1.set_title('Tn(minimum Temperature)')
ax1.set_xlabel('Month')
ax1.set_ylabel('p')



ax1.plot(month,df_accra_vec_tn.p,marker='o')
ax1.plot(month,df_nav_vec_tn.p,marker='o')
ax1.plot(month,df_tem_vec_tn.p,marker='o')
ax1.plot(month,df_kum_vec_tn.p,marker='o')

# A plot of  Tm for each region 
ax2.set_title('Tm(maximum Temperature)')
ax2.plot(month,df_accra_vec_tm.p,marker='o')
ax2.plot(month,df_nav_vec_tm.p,marker='o')
ax2.plot(month,df_tem_vec_tm.p,marker='o')
ax2.plot(month,df_kum_vec_tm.p,marker='o')
ax2.set_xlabel('Month')
ax2.set_ylabel('p')

# A plot of  Tx for each region 
ax3.set_title('Tx(mean temperature)')
ax3.plot(month,df_accra_vec_tx.p,marker='o')
ax3.plot(month,df_nav_vec_tx.p,marker='o')
ax3.plot(month,df_tem_vec_tx.p,marker='o')
ax3.plot(month,df_kum_vec_tx.p,marker='o')
ax3.set_xlabel('Month')
ax3.set_ylabel('p')

# Setting grid sizes 
ax1.grid(which="minor",alpha=0.3,color='#000')
ax1.grid(which="major",alpha=0.6,color='#000')
ax1.minorticks_on()

ax2.grid(which="minor",alpha=0.3,color='#000')
ax2.grid(which="major",alpha=0.6,color='#000')
ax2.minorticks_on()

ax3.grid(which="minor",alpha=0.3,color='#000')
ax3.grid(which="major",alpha=0.6,color='#000')
ax3.minorticks_on()

fig.legend(labels, loc = 'lower right')
plt.legend()
#print('Fig 1. Monthly Survival Rate of Mosquito ')
plt.savefig('output_graphs/rate.jpeg')


plt.show()

```
